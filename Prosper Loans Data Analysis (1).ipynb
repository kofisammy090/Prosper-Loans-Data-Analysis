# Question One
How's the distribution of the borrower apr in the dataset, do the majority of loans have high or low APR values?

# Question six
Whats the relationship between Borrower apr and loan amount

# Visualization

```python
plt.figure(figsize=[8,5])
sb.regplot(data=df,y='borrower apr', x='loanoriginalamount',scatter_kws={'alpha':0.01})
plt.title('Borrower Apr vrs Loan Amount', size=14)
plt.xlabel('Loan Original Amount',size=14)
plt.ylabel('Borrower Apr', size=14);
```

# Observation

NB.
We can see there is a negative relationship between borrower apr and loan amount. At different size of the loan amount, the APR has a large range, but the range of APR decrease with the increase of loan amount. This means as the borrower APR decreases the loan amount also increases. The higher the loan amount the lower the interest rate

# Observation
 We notice the borrower APR decreases while we have the rating increasing. Borrowers with the best Prosper ratings have the lowest APR. It means that the Prosper rating has a strong effect on borrower APR

# Question Ten
Whats the relationship between loan amount and Term

```python
plt.figure(figsize=[8,5])
sb.regplot(data=df,x='term',y='loanoriginalamount',scatter_kws={'alpha':0.01}, x_jitter=4, y_jitter=500)
plt.title('Loan Amount and Loan Term',size=14)
plt.xlabel('Months',size=14)
plt.ylabel('Loan Amount',size=14);
```

# Observation
There is a positive relationship between the loan amoun and loan term. The loan term increase as the loan amount increase. The darker part of the of plot shows that most of the borrowers prefer 36 months loan term.

# Multi-variate

# Question Eleven

Whats the relationship between borrower apr, prosper rating and term?

# Visualization

```python
plt.figure(figsize=[8,5])
sb.pointplot(data = df, y='borrower apr', x= 'prosperrating (numeric)', hue='term',linestyles='',dodge=0.4,ci='sd', color='red')
plt.title('Borrower apr by Prosper Rating and Term')
plt.xlabel('Prosper Rating', size= 14)
plt.ylabel('Borrower Apr', size=14);
```

# Observation
As seen the borrower APR decrease with the increase of borrow term for people with 1.0-4.0 ratings. But for people with 5.0-7.0 ratings, the APR increase with the increase of borrow term.
